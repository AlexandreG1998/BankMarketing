```python
# Import numpy and pandas to prepare dataset; LabelEncoder to encoding data;

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import random
```

```python
dataset = pd.read_csv('bank-full.csv', sep=';')
dataset.head()
```

```python
#Re-name colum names for better visualization

dataset.columns = ['Age','Job','Marital','Education','Default','Balance','Housing','Loan','Contact',
                   'Day','Month','Duration','Campaign','Pdays','Previus','Poutcome','Y']

dataset.head()
```

```python
df = dataset.dropna()
jobencoder = LabelEncoder()
maritialencoder = LabelEncoder()
educationencoder = LabelEncoder()
defaultencoder = LabelEncoder()
housingencoder = LabelEncoder()
loanencoder = LabelEncoder()
contactencoder = LabelEncoder()
monthencoder = LabelEncoder()
poutcomeencoder = LabelEncoder()

df.iloc[:,1:2] = jobencoder.fit_transform(df.iloc[:,1:2].values)

df.iloc[:,2:3] = maritialencoder.fit_transform(df.iloc[:,2:3].values)

df.iloc[:,3:4] = educationencoder.fit_transform(df.iloc[:,3:4].values)

df.iloc[:,4:5] = defaultencoder.fit_transform(df.iloc[:,4:5].values)


df.iloc[:,6:7] = housingencoder.fit_transform(df.iloc[:,6:7].values)

df.iloc[:,7:8] = loanencoder.fit_transform(df.iloc[:,7:8].values)

df.iloc[:,8:9] = contactencoder.fit_transform(df.iloc[:,8:9].values)

df.iloc[:,10:11] = monthencoder.fit_transform(df.iloc[:,10:11].values)

df.iloc[:,15:16] = poutcomeencoder.fit_transform(df.iloc[:,15:16].values)


df.head()
```

```python
df.iloc[:,16:17] = np.where(df.iloc[:,16:17] == 'yes',1,0)
df.head()
```

```python
dfB = df.values
dfB
```

```python
dataone = df
dataone.head()
```

```python
k = dataone.drop(dataone[dataone["Y"] == 0].index,  inplace=True)
```

```python
dataone.head()
```

```python
dfB = pd.DataFrame(dfB)
dfB.columns = dataone.columns

dfB.head()
```

```python
datazeros = dfB
k = datazeros.drop(datazeros[datazeros["Y"] == 1].index)
```

```python
number_of_train_rows = int(len(dataone) * 0.5)
    
train_one = dataone[:number_of_train_rows]
train_zero = k[:number_of_train_rows]
    
train_data = pd.concat((train_one,train_zero), axis = 0)
    
test_one = dataone[number_of_train_rows:]
test_zero = k[number_of_train_rows:]
    
test_data = pd.concat((test_one,test_zero), axis = 0)
```

```python
train_y = train_data.iloc[:,16:].values

train_x = train_data.drop('Y', axis = 1).values

test_y = test_data.iloc[:,16:].values
test_x = test_data.drop('Y', axis = 1).values
```

```python
# Import Keras

from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
model = Sequential()
model.add(Dense(units = train_x.shape[1], input_dim = train_x.shape[1], activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(units = 64, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(units=1, activation = 'sigmoid'))
model.compile(optimizer = "adam", loss = "binary_crossentropy", metrics=["acc"])

history = model.fit(train_x, train_y, batch_size=1000, epochs=50)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history["acc"])
plt.title('Accuracy step by step')
plt.show()
```

```python
model.evaluate(test_x, test_y, batch_size=1000)
```
